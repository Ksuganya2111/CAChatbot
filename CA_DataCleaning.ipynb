```python
with open("/content/jobs.txt", "r") as f:
  dirty = f.readlines()
```

```python
print(dirty)
```

```python
clean_jobs = {}
```

```python
for count,elem in enumerate(dirty):
  if elem.startswith("###ID"):
    ID = elem[3:].strip()
    clean_jobs[ID] = {}
    clean_jobs[ID]["TITLE"] = ""
    clean_jobs[ID]["RAW"] = []
    # print(dirty[count+1])
    # clean_jobs[ID]["Title"] = ".".split(dirty[i+1])[0]
  elif elem.startswith("###TITLE"):
    TITLE = elem[8:-1]
    clean_jobs[ID]["TITLE"] = TITLE
  elif elem.startswith("###COMPANYNAME"):
    COMPANY_NAME = elem[14:-1]
    clean_jobs[ID]["COMPANY_NAME"] = COMPANY_NAME
  elif elem.startswith("###LOCATION"):
    LOCATION = elem[11:-1]
    clean_jobs[ID]["LOCATION"] = LOCATION
  elif elem.startswith(TITLE) or elem == "\n" or elem.startswith("save-icon"):
    continue
  else:
    clean_jobs[ID]["RAW"].append(elem.strip())
  
```
